```python
import pandas as pd
import json as j
from pandas.io.json import json_normalize
import nltk
import math
import re
```

```python
with open('C:\wamp64\www\youtube comments grabber\lions.json') as comment_json:
    data = j.load(comment_json)
    d = data
    my_list= []
    out =[]
    users=[]
    valid_words =[]
    validcomments=[]
def valid(t) :
        p = re.compile('[a-z]*')
        t= nltk.word_tokenize(t)
        
        ss=[]
       
        for words in t:
                   
                    w = p.match(words)
                    e = w.group()
                    #print(e)
                    ss.append(e)
                   
                   
            

        return(ss)
             
```

```python
for x in d:
        a = x['comm']
        b= x['username']
        users.append(b)

        my_list.append(a)
from nltk.corpus import stopwords

stopword =['','br','href','http']
for text in my_list:
        text = text.lower() 
        out = valid(text)
        
        for i in out :
            if i == None :
                
                out.remove(i)
        out =[word for word in out if word not in stopword]    
        valid_words.append(out)
        out = ' '.join(out)
        validcomments.append(out)
        
        
print(valid_words)
print(validcomments)
```

```python
def cosine_similarity(vector1, vector2):
    dot_product = sum(p*q for p,q in zip(vector1, vector2))
    magnitude = math.sqrt(sum([val**2 for val in vector1])) * math.sqrt(sum([val**2 for val in vector2]))
    if not magnitude:
        return 0
    return dot_product/magnitude
def wordsbag(c,sub) :
    return(list(set().union(sub,c)))
def create_vectors(wordlist,cmnt):
    vector=[]
    for i in wordlist:
        if i in cmnt:
            vector.append(cmnt.count(i))
        else:
            vector.append(0)
    return(vector)      
```

```python
sub = valid_words[0]
uinque_wordbag=[]
similarity =[]
for ls in valid_words:
    unique_wordbag =wordsbag(ls,sub)
    
    similarity.append(cosine_similarity(create_vectors(unique_wordbag ,ls),create_vectors(unique_wordbag ,sub) ))
    
    
print(similarity)    
```

```python
df = pd.DataFrame(validcomments, columns = ['sentence'])
df
```

```python
df =df.assign(username =users, similarity_with_sub =similarity)
```
